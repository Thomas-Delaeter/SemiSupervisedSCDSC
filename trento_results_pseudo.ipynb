```python
labels = ['baseline', '\u03C4=80%', '\u03C4=70%', '\u03C4=60%']

accuracy = [0.8828, 0.8772, 0.8878, 0.8878]
nmi = [0.9081, 0.8898, 0.8995, 0.8996]
kappa = [0.8434, 0.8357, 0.8497, 0.8497]
misses = [400, 3, 0, 0]

trento03 = np.vstack((accuracy, nmi, kappa))
trento03

accuracy = [0.8828, 0.8876, 0.8876, 0.7813]
nmi = [0.9081, 0.8983, 0.8985, 0.7899]
kappa = [0.8434, 0.8494, 0.8494, 0.6961]
misses = [400, 3, 0, 0]

trento05 = np.vstack((accuracy, nmi, kappa))
trento05
```
